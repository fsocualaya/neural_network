```python
import numpy as np
import matplotlib.pyplot as plt
import csv 
import pandas as pd
```

```python
dfi = pd.read_csv('iris.data',header=None,names=['sepal_length_cm','sepal_width_cm','petal_length_cm','petal_width_cm','class'])
```

```python
dfi
```

```python
# PRE-PROCESSING

dfi1 = dfi[dfi['class']!='Iris-virginica'] # Eliminamos la especie iris-virginica
dfi1 = dfi1[["petal_length_cm","petal_width_cm","class"]] # Elejimos las variables que nos importan

# Tambien vamos a cambiar la especie de palabras a datos categoricos para que los algoritmos corran mas rapido
dfi1['class'].replace({"Iris-setosa": 0, "Iris-versicolor": 1}, inplace=True)

dfi1
```

```python
groups = dfi1.groupby('class')

for name, group in groups:
  plt.plot(group["petal_length_cm"], group["petal_width_cm"], marker="o", linestyle="", label=name)
  #pueden cambiar los variables de las ejes X y Y cambiando los inputs de plt.plot() justo arriba ^

  plt.legend()
plt.show()
```

```python
dfi1.to_csv('Iris-reducido.csv')
```

```python
dfe = pd.read_csv('mnr_8651_sm_table1.txt', header = None, names = ['ASAS ID', "Imag", 'std_error', 'num_measure', 'period', 'amplitude', 'R21', 'phase_dif', 'num_harmonics', "class", "prob"], sep = '\s+')
dfe
```

```python
groups = dfe.groupby('class')

for name, group in groups:
    plt.plot(group["period"], group["amplitude"], marker="o", linestyle="", label=name)
    #pueden cambiar los variables de las ejes X y Y cambiando los inputs de plt.plot() justo arriba ^
    
    plt.legend()
    plt.xscale('log') # Cuando pongan el periodo en un grafo, les recomendo que este en escala logaritmica
    plt.yscale('log') # Cuando pongan la amplitud en un grafo, les recomendo que este en escala logaritmica
    # Para R21 y phase_dif, mantenerlo en escala lineal quitando la linea correspondiente de arriba

plt.show()
```

```python
# PRE-PROCESSING

dfe1 = dfe[['period', "amplitude", "R21", "phase_dif", "class"]]
dfe1.dropna(inplace=True)

dfe1.loc[dfe1['class'] == 4, 'class'] = 0 # Juntamos las clases 0, 3, y 4 en una sola clase (0)
dfe1.loc[dfe1['class'] == 3, 'class'] = 0 # Juntamos las clases 0, 3, y 4 en una sola clase (0)
dfe1.loc[dfe1['class'] != 0, 'class'] = 1 # Juntamos el resto de las clases en la clase 1
# Ahora las clases son: 0 = binaria ecliptica, 1 = no binaria ecliptica

# Si les sale un warning, ignorenlo y continuen
```

```python
dfe1
```

```python
groups = dfe1.groupby('class')

for name, group in groups:
    plt.plot(group["period"], group["amplitude"], marker="o", linestyle="", label=name)
    #pueden cambiar los variables de las ejes X y Y cambiando los inputs de plt.plot() justo arriba ^
    
    plt.legend()
    plt.xscale('log') # Cuando pongan el periodo en un grafo, les recomendo que este en escala logaritmica
    plt.yscale('log') # Cuando pongan la amplitud en un grafo, les recomendo que este en escala logaritmica
    # Para R21 y phase_dif, mantenerlo en escala lineal quitando la linea correspondiente de arriba

plt.show()
```
